```python
import torch
from adapters.pip_net import QuantizedPIPNetHead
```

### Set shapes

```python
NUM_CLASSES: int = 4
NUM_PROTOS:  int = 100
PROTO_DIM:   int = 768

BATCH_SIZE:  int = 3
FMAP_SIZE:   int = 7
```

### Mock codebook

```python
codebook = torch.randn((NUM_PROTOS, PROTO_DIM))
```

### Init PIP-Net-like head with codebook

```python
model = QuantizedPIPNetHead(
    num_classes=NUM_CLASSES,
    codebook=codebook, 
    normalize=True,
    bias=False,
)
```

### Mock featuremap (backbone output)

```python
x = torch.randn(
    (BATCH_SIZE, PROTO_DIM, FMAP_SIZE, FMAP_SIZE), 
)
```

### Predict

```python
result = model(x)
```

### Check result

```python
assert result.proto_fmap.shape == (BATCH_SIZE, NUM_PROTOS, FMAP_SIZE, FMAP_SIZE)
```

```python
assert result.proto_fvec.shape == (BATCH_SIZE, NUM_PROTOS)
```

```python
assert result.logits.shape == (BATCH_SIZE, NUM_CLASSES)
```

```python
assert result.pre_softmax.shape == (BATCH_SIZE, NUM_CLASSES)
```

```python
assert result.softmax_out.shape == (BATCH_SIZE, NUM_CLASSES)
```
